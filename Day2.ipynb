```
firstInput = int(input("Enter a number: "))

print("Given input is", firstInput)

add = firstInput + 10
sub = firstInput - 10
mul = firstInput * 10
div = firstInput / 10

print("Adding with 10 gives:", add)
print("Subtracting with 10 gives:", sub)
print("Multiplying with 10 gives:", mul)
print("Dividing with 10 gives:", div)
```

```
firstInput = int(input("Enter first number: "))
secondInput = int(input("Enter second number: "))

print("Given inputs are", firstInput, secondInput)

add = firstInput + secondInput
sub = firstInput - secondInput
mul = firstInput * secondInput
div = firstInput / secondInput

print("Adding both numbers give:", add)
print("Subtracting both numbers give:", sub)
print("Multiplying both numbers gives:", mul)
print("Dividing both numbers gives:", div)
```
